```python
import pandas as pd
import json
```

```python
fv = pd.read_csv("feature_values37.csv", names=["feature","value"])
```

```python
fv
```

```python
with open("../feature_dict.json") as f:
    feature_dict = json.load(f)
```

```python
feature_dict
```

```python
fd = pd.DataFrame(feature_dict, index=[
    "feature_weight"
]).transpose().reset_index().rename(columns={"index": "feature"})
```

```python
fd
```

```python
complete_dict = pd.merge(fv, fd, "outer")
```

```python
fixed = complete_dict.fillna(1)
```

```python
fixed['weighted'] = fixed.apply(lambda x: x.value * x.feature_weight, axis="columns")
```
